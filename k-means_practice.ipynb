```python
import pickle
import matplotlib.pyplot as plt
import sys
import sklearn as sk
import numpy 
import pylab
from sklearn.cluster import KMeans
```

```python
X=numpy.genfromtxt('clusterData.csv',delimiter=',', skip_header=0)
```

```python
plt.scatter(X[:,0],X[:,1])
initial=numpy.array([[-2,0],[0,6], [1,2]]) # 3 cluster centers
plt.scatter(initial[:,0], initial[:,1], s=100, marker='^')
pylab.savefig("cl3_0.png")
plt.show()
```

```python
for i in numpy.arange(1,6):
    kmeans = KMeans(n_clusters=3, init= initial, max_iter=i, n_init=1, random_state=0)
    kmeans.fit(X)
    pred=kmeans.predict(X)
    plt.scatter(X[:,0],X[:,1],c=pred)
    plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1] , c="r", s=100, marker='^')
    n="cl3_"+str(i)+".png"
    pylab.savefig(n)
    plt.show()
    
```

```python
plt.scatter(X[:,0],X[:,1])
clustNo=4
########### Making 4 cluster centers ###########
a=numpy.random.uniform(-4,4,clustNo)
b=numpy.random.uniform(-2,6,clustNo)
initial=numpy.stack((a, b), axis=-1)
##################### 'initial' contains 4 initial cluster centers  ######
plt.scatter(initial[:,0], initial[:,1], s=100, marker='^')
pylab.savefig("cl4_0.png")
plt.show()
```

```python
for i in numpy.arange(1,10):
    kmeans = KMeans(n_clusters=4, init= initial, max_iter=i, n_init=1, random_state=0)
    kmeans.fit(X)
    pred=kmeans.predict(X)
    plt.scatter(X[:,0],X[:,1],c=pred)
    plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1] , c="r", s=100, marker='^')
    n="cl4_"+str(i)+".png"
    pylab.savefig(n)
    plt.show()
    
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)      
```

```python
h=.01
x_min=-3
x_max=4
y_min=-1
y_max=10
centers=kmeans.cluster_centers_
import numpy as np
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z =kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z) #play around with cmap options to get different color maps
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7)
plt.show()
```
