# SVM non linear

```python
import numpy as np
import sklearn as sk
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data=np.genfromtxt("non-linearly-separable.csv")
```

```python
data.shape

```

```python
X=data[:,range(2)]
```

```python
y=data[:,2]
```

```python
plt.scatter(X[:,0], X[:,1],c=y,s=70,cmap='winter')
```

```python
from sklearn.svm import SVC
```

```python
# We use Radial Basis Function, or rbf to transform the data
clf=SVC(kernel='rbf', C=1)
```

```python
clf.fit(X,y)
```

```python
np.sum(clf.predict(X)!=y)
```

```python
h=.01
x_min=-1.5
x_max=1.5
y_min=-1.5
y_max=1.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z) #play around with cmap options to get different color maps
plt.scatter(X[:,0], X[:,1],c=y,s=70,cmap='winter')
plt.show()
```
