```python
import numpy as np
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
import math

```

```python
d=np.genfromtxt('clusterData.csv', delimiter=',', skip_header=1)
```

```python
sns.scatterplot(d[:,0], d[:,1],legend=False)
```

```python
from sklearn.cluster import KMeans
```

```python
X=d[:,np.arange(2)]
```

```python
X.shape
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

# plot with given values [ -2,0],[ 0,6],[1,2]

```python
kmeans = KMeans(n_clusters=3, init=np.array([[ -2,0],[ 0,6],[1,2]]), random_state=0, max_iter=1).fit(X)
centers = kmeans.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7)
```

# plot with intial clusters

# Experiment with different initial cluster centers, for k=3. Do the cluster centers always end up at the same location regardless of initial positions?


```python
kmeans = KMeans(n_clusters=3, init=np.array([[ -2,0],[ 0,6],[1,2]]), random_state=0, max_iter=1).fit(X)
centers = kmeans.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7)
```
